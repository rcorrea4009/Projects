```python
numeric_cols = df.select_dtypes(include=["int", "float"]).columns.tolist()
target = "heart_disease_present"

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=target, y=col, data=df, palette="Set3")
    plt.title(f'{col} by Target Class')
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int", "float"]).columns.tolist()

outliers = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
  
    outlier_rows = df[(df[col] < low) | (df[col] > high)]
    outliers[col] = outlier_rows.shape[0]  

outliers
```
